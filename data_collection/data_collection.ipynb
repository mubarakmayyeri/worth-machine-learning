### Data Cleaning

```python
# Removing unnecessry columns
data = gform.iloc[:, 2:]
data.head()
```

```python
# Renaming column names
cols = ['age', 'domain', 'current_job', 'degree', 'cs_bg', 'experience', 'avg_time', 'holidays', 'week_back', 'city', 'skills', 'graduated', 'toi', 'salary']
data.columns = cols
data.head()
```
