# Model Training - Combined dataset

Models to train
1. Linear regression
2. Polynomial regression
3. KNN
4. Random Forest

```python
# Importing necessary libraries
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
# Reading the datasets
X_train = pd.read_csv('data/combined_dataset/X_train.csv')
X_test = pd.read_csv('data/combined_dataset/X_test.csv')
y_train = pd.read_csv('data/combined_dataset/y_train.csv')
y_test = pd.read_csv('data/combined_dataset/y_test.csv')
```

## Model training

### 1. Linear regression

```python
# Training the model with training dataset
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
# Evaluating the performance of the model
from sklearn.metrics import mean_squared_error, r2_score

y_pred = lr.predict(X_test)

# Computing evaluation metrics for the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error for Linear regressor: {mse:.2f}")
print(f"R^2 Score for Linear regressor: {r2:.2f}")
```

```python
# Creating a Dataframe for stroring model performance
#storing evaluation metrics to a dataframe

perfo = pd.DataFrame([['Linear Regression', f'{mse:.2f}', f'{r2:.2f}']],
                       columns=['Model', 'MSE', 'R2 Error'])

perfo
```

### 2. Polynomial Regression 

```python
polynomial_features = PolynomialFeatures(degree=2)

X_train_poly = polynomial_features.fit_transform(X_train)
X_test_poly = polynomial_features.transform(X_test)
```

```python
X_train_poly.shape
```

```python
# Training the model
pr = LinearRegression()
pr.fit(X_train_poly, y_train)
```

```python
# Evaluating the performance of the mbodel
from sklearn.metrics import mean_squared_error, r2_score

y_pred = pr.predict(X_test_poly)

# Computing evaluation metrics for the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error for Polynomial regressor: {mse:.2f}")
print(f"R^2 Score for Polynomial regressor: {r2:.2f}")
```

```python
#storing evaluation metrics to a dataframe

pr_results = pd.DataFrame([['Polynomial Regression', f'{mse:.2f}', f'{r2:.2f}']],
                       columns=['Model', 'MSE', 'R2 Error'])

perfo = perfo.append(pr_results, ignore_index=True)
perfo
```

### 3. K Neighbors Regressor

```python
# Training the model
knn = KNeighborsRegressor(n_neighbors=5)

knn.fit(X_train, y_train)
```

```python
# Evaluating the performance of the model
from sklearn.metrics import mean_squared_error, r2_score

y_pred = knn.predict(X_test)

# Computing evaluation metrics for the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error for KNN regressor: {mse:.2f}")
print(f"R^2 Score for KNN Polynomial: {r2:.2f}")
```

#### Hyperparmeter tuning

```python
# finding best value for "k"
from sklearn.model_selection import GridSearchCV

param_grid = {'n_neighbors': [i for i in range(1, 25)]}
grid_search = GridSearchCV(estimator=knn, param_grid=param_grid, cv=5)
grid_search.fit(X_train, y_train)

best_n_neighbors = grid_search.best_params_['n_neighbors']
best_n_neighbors
```

```python
# Training KNN with best value for k
final_knn = KNeighborsRegressor(n_neighbors=best_n_neighbors)
final_knn.fit(X_train, y_train)
```

```python
# Evaluating the performance of the model
from sklearn.metrics import mean_squared_error, r2_score

y_pred = final_knn.predict(X_test)

# Computing evaluation metrics for the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error for improved KNN regressor: {mse:.2f}")
print(f"R^2 Score for improved KNN Polynomial: {r2:.2f}")
```

```python
#storing evaluation metrics to a dataframe

knn_results = pd.DataFrame([['KNN Regressor', f'{mse:.2f}', f'{r2:.2f}']],
                       columns=['Model', 'MSE', 'R2 Error'])

perfo = perfo.append(knn_results, ignore_index=True)
perfo
```

### Random Forest Regressor

```python
# Random Forest
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor()
rf.fit(X_train, np.ravel(y_train))
```

```python
y_pred = rf.predict(X_test)

# Computing evaluation metrics for the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error for RF regressor:", mse)
print("R^2 Score for RF regressor:", r2)
```
